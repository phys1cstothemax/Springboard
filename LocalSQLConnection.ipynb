```python
import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn
```

```python
## Q1: Some of the facilities charge a fee to members, but some do not.
## Write a SQL query to produce a list of the names of the facilities that do. 
```

```python
def select_all_tasks(conn):
    cur = conn.cursor()
    
    query1 = """
    SELECT name
    FROM `Facilities`
    WHERE membercost = 0
        
        """
    cur.execute(query1)

    rows = cur.fetchall()
 
    for row in rows:
        print(row)
        
def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("Query 1 results:")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
## Q5: Produce a list of facilities, with each labelled as'cheap' or 'expensive',
## depending on if their monthly maintenance cost is
## more than $100. Return the name and monthly maintenance of the facilities
## in question.
```

```python
def select_all_tasks(conn):
    cur = conn.cursor()
    
    query5 = """
    SELECT name, monthlymaintenance, 
    CASE WHEN monthlymaintenance > 100 
         THEN 'expensive'
         ELSE 'cheap'
         END AS maintenancecost
    FROM `Facilities`
        
        """
    cur.execute(query5)

    rows = cur.fetchall()
 
    for row in rows:
        print(row)
        
def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("Query 5 results:")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
## Q7: Produce a list of all members who have used a tennis court.
## Include in your output the name of the court, and the name of the member
## formatted as a single column. Ensure no duplicate data, and order by
## the member name.
```

```python
def select_all_tasks(conn):
    cur = conn.cursor()
    
    query7 = """
    SELECT DISTINCT b.facid AS Facility_ID, f.name AS Facility, b.memid AS Member_ID, 
        (m.firstname || ' ' || m.surname) AS Member 
    FROM `Members` AS m 
    INNER JOIN `Bookings` AS b
    ON m.memid = b.memid 
    INNER JOIN `Facilities` AS f
    ON f.facid = b.facid 
    WHERE b.facid IN (0, 1) AND b.memid != 0
    ORDER BY Member;
        """
    cur.execute(query7)

    rows = cur.fetchall()
 
    for row in rows:
        print(row)
        
def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("Query 7 results:")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
## Q8: Produce a list of bookings on the day of 2012-09-14 which
## will cost the member (or guest) more than $30. Remember that guests have
## different costs to members (the listed costs are per half-hour 'slot'), and
## the guest user's ID is always 0. Include in your output the name of the
## facility, the name of the member formatted as a single column, and the cost.
## Order by descending cost, and do not use any subqueries.
```

```python
def select_all_tasks(conn):
    cur = conn.cursor()
    
    SQLquery8 = """
    SELECT f.name, (m.firstname||' '||m.surname) AS name, CASE WHEN b.memid != 0
    THEN f.guestcost * b.slots
    ELSE f.membercost * b.slots
    END AS cost
    FROM Bookings AS b
    INNER JOIN Facilities AS f ON b.facid = f.facid
    AND b.starttime LIKE  '2012-09-14%'
    AND (((b.memid =0) AND (f.guestcost * b.slots >30))
    OR ((b.memid !=0) AND (f.membercost * b.slots >30)))
    INNER JOIN Members AS m ON b.memid = b.memid
    ORDER BY cost DESC
        
        """
    cur.execute(SQLquery8)

    rows = cur.fetchall()
         
    for row in rows:
        print(row)
        
def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("Query 8 results:")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
## Q9: This time, produce the same result as in Q8, but using a subquery.
```

```python
def select_all_tasks(conn):
    cur = conn.cursor()
    
    query9 = """
    SELECT sub.starttime, sub.name, 
    CASE WHEN sub.memid != 0 THEN (sub.firstname || ' ' || sub.surname)
        ELSE sub.firstname 
    END AS name, 
    CASE WHEN sub.memid != 0 THEN sub.slots * sub.membercost
        ELSE sub.slots * sub.guestcost
    END AS cost
    FROM (select b.starttime, b.facid, b.memid, b.slots, f.membercost, f.guestcost, f.name,
        m.firstname, m.surname
        FROM Bookings AS b
        INNER JOIN Members AS m
        ON b.memid = m.memid
        INNER JOIN Facilities AS f
        ON f.facid = b.facid) AS sub
    WHERE CAST(starttime AS date) = CAST('2012-09-14' AS date) AND
    CASE WHEN memid != 0 THEN slots * membercost
        ELSE slots * guestcost
    END > 30
    ORDER BY cost DESC;
        """
    cur.execute(query9)

    rows = cur.fetchall()
 
    for row in rows:
        print(row)
        
def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("Query 9 results:")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
## Q10: Produce a list of facilities with a total revenue less than 1000.
## The output of facility name and total revenue, sorted by revenue. Remember
## that there's a different cost for guests and members!
```

```python
import sqlite3 
import pandas as pd 

con = sqlite3.connect("sqlite_db_pythonsqlite.db")
cur = con.cursor()
```

```python
query_10 = '''
SELECT name, SUM(rev) AS total_revenue 
FROM (SELECT f.name, CASE WHEN b.memid != 0 
                     THEN f.membercost*b.slots 
                     ELSE f.guestcost*b.slots END AS rev
      FROM Facilities AS f 
     INNER JOIN Bookings AS b
ON b.facid = f.facid) AS sub
GROUP BY sub.name
HAVING SUM(rev) < 1000
ORDER BY total_revenue;
'''
cur.execute(query_10)

print("{0:10s} \t {1:10s}".format('Facility', 'Total Rev'))
for res in cur.fetchall():
    print("{0:10s} \t {1:.2f}".format(*res))
```

```python
## Q13: Find the facilities usage by month, but not guests
```

```python
query_13 = '''
SELECT f.name, (strftime('%Y-%m', starttime)) AS month_count, count(*) AS usage_count
FROM Facilities AS f
INNER JOIN Bookings AS b
ON b.facid = f.facid
WHERE b.memid != 0
GROUP BY f.name, month_count;
'''
cur.execute(query_13)

print("{0:15s} \t {1:10s} \t {2:10s}".format('Facility', 'Month', 'Usage'))
for res in cur.fetchall():
    print("{0:15s} \t {1:10s} \t {2:.2f}".format(*res))
```
